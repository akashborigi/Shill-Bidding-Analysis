```python
import pandas as pd
import numpy as np
from numpy import mean
from numpy import absolute
import torch
import torch.nn as nn
import torch.nn.functional as F
from matplotlib import pyplot as plt
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from skelm import ELMClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
import time
from sklearn.metrics import mean_squared_error
from vecstack import stacking
```

```python
data=pd.read_csv("Shill Bidding Dataset.csv")
data=pd.DataFrame(data)
X=np.array(data.iloc[0:,3:12])
Y=np.array(data.iloc[0:,12:])
```

```python
LE=LabelEncoder()
X[:,0]=LE.fit_transform(X[:,0]) #Select labels from variables
#Split train and test data. Test data size is 50%, training set is 50%
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.5, random_state = 0) 
#The data set is normalized or 'standardized' for a better performaing KNN for both train and test data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Set up the pipeline
svm_pipe = make_pipeline(StandardScaler(), SVC())

# Set up the grid search
param_grid = {
    'svc__C': [0.1, 1, 10],
    'svc__kernel': ['poly', 'rbf', 'sigmoid'],
    'svc__gamma': ['scale', 'auto']
}
clf = GridSearchCV(svm_pipe, param_grid, cv=5)

# Fit the model and time it
clf.fit(X_train, y_train)

# Use the best parameters to fit a new SVM model
best_params = clf.best_params_
svm_pipe = make_pipeline(SVC(C=best_params['svc__C'], kernel=best_params['svc__kernel'], gamma=best_params['svc__gamma']))
```

```python
#Linear SVM
linSVM = LinearSVC(penalty='l2', dual=False, max_iter=10000)
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}

start_time = time.time()
# Set up k-fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
# Set up GridSearchCV to perform model selection using k-fold cross-validation
clf = GridSearchCV(linSVM, param_grid=param_grid)
clf.fit(X_train, y_train)
end_time = time.time()
duration1=end_time-start_time
# Get the best model
linSVM_model = clf.best_estimator_
print(f"Best model: {linSVM_model}")
```

```python
#Create and initialize deep learning classifier
gaussDL = GaussianNB()
```

```python
# putting all base model objects in one list
all_models = [svm_pipe,linSVM, gaussDL]
```

```python
# computing the stack features
s_train, s_test = stacking(all_models, X_train, y_train, X_test,regression=True, n_folds=4,shuffle=True)
 
# initializing the second-level model
final_model = svm_pipe
 
# fitting the second level model with stack features
final_model = final_model.fit(s_train, y_train)
 
# predicting the final output using stacking
y_pred = final_model.predict(s_test)

print('Mean Squared Error: ', mean_squared_error(y_test, y_pred))

print(classification_report(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_pred,y_test),'\n')
print('Accuracy:', round(accuracy_score(y_pred,y_test)*100,2),'%')
```
